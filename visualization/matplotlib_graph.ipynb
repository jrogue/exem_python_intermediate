```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

budget = pd.read_csv("https://raw.githubusercontent.com/jrogue/exem_python_intermediate/master/data/mn-budget-detail-2014.csv")
budget = budget.sort_values('amount',ascending=False)[:10]

details = budget["detail"].values
amounts = budget["amount"].values
 
y_pos = np.arange(len(details))
 
plt.bar(y_pos, amounts, align='center', alpha=0.5)
plt.xticks(y_pos, details)
plt.xticks(rotation=90)
plt.ylabel('amount')
plt.title('MN Capital Budget - 2014')
 
plt.show()
```
