```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

budget = pd.read_csv("https://raw.githubusercontent.com/jrogue/exem_python_intermediate/master/data/mn-budget-detail-2014.csv")
budget = budget.sort_values('amount',ascending=False)[:10]
```

```python
sns.set_style("darkgrid")
bar_plot = sns.barplot(x=budget["detail"],y=budget["amount"],
                        palette="muted")
plt.xticks(rotation=90)
plt.show()
```
